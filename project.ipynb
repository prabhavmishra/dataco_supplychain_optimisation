```python
import pandas as pd
```

```python
df=pd.read_csv('dataco_data.csv', encoding='latin-1')
```

```python
cols_to_keep = [col for col in df.columns if col in [
    'Type', 'Days for shipping (real)', 'Days for shipment (scheduled)',
    'Benefit per order', 'Category Name', 'Customer City', 'Customer Segment',
    'Order City', 'Order Country', 'order date (DateOrders)', 'Order Region', 
    'Order Status', 'Product Name', 'Sales', 'Shipping Mode', 'Late_delivery_risk'
]]
df = df[cols_to_keep]
df.head()
```

```python
df = df.drop_duplicates()
```

```python
df.columns = [col.replace(' ', '_').replace('(', '').replace(')', '').lower() for col in df.columns]
```

```python
df['order_date_dateorders'] = pd.to_datetime(df['order_date_dateorders'])
```

#creating kpi: delay days


```python
df['delay_days'] = df['days_for_shipping_real'] - df['days_for_shipment_scheduled']
```

```python
print(f"Final Row Count: {len(df)}")
```

```python
# Check for duplicates and null values after cleaning
print("=" * 50)
print("DATA QUALITY CHECK AFTER CLEANING")
print("=" * 50)
print(f"\nDataset Shape: {df.shape}")
print(f"Total Rows: {df.shape[0]}")
print(f"Total Columns: {df.shape[1]}")
print(f"\nDuplicate Rows: {df.duplicated().sum()}")
print(f"\nNull Values by Column:")
null_summary = df.isnull().sum()
if null_summary.sum() == 0:
    print("  ✓ No null values found in the dataset")
else:
    print(null_summary[null_summary > 0])
print("\n" + "=" * 50)
```

```python
df.to_csv('dataco_cleaned.csv', index=False)
```
